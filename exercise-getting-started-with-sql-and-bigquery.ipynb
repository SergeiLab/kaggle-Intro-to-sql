### 3) Create a crime map

If you wanted to create a map with a dot at the location of each crime, what are the names of the two fields you likely need to pull out of the `crime` table to plot the crimes on a map?

```python
# Write the code here to explore the data so you can find the answer
# --- 1. Construct a reference to the 'crime' table ---
table_ref = dataset_ref.table("crime")

# --- 2. API request - fetch the table ---
table = client.get_table(table_ref)

# --- 3. Explore the table schema to find fields with relevant names/types ---
print("Schema Fields:")
# You'd typically look for fields with names like 'latitude', 'longitude', or data types like 'FLOAT' or 'GEOGRAPHY'
relevant_fields = []
for field in table.schema:
    if 'LATITUDE' in field.name.upper() or 'LONGITUDE' in field.name.upper() or 'LOCATION' in field.name.upper():
        relevant_fields.append((field.name, field.field_type))

# Print the names of the most relevant fields
print("\nFields relevant for plotting (Latitude/Longitude):")
for name, field_type in relevant_fields:
    print(f"- {name} (Type: {field_type})")
```

Thinking about the question above, there are a few columns that appear to have geographic data. Look at a few values (with the `list_rows()` command) to see if you can determine their relationship.  Two columns will still be hard to interpret. But it should be obvious how the `location` column relates to `latitude` and `longitude`.

```python
# Scratch space for your code

# Fetch the first 5 rows, but only select the relevant geographic columns
# We also include 'unique_key' to ensure each row is distinct
query = f"""
SELECT
  unique_key,
  latitude,
  longitude,
  location,
  x_coordinate,
  y_coordinate
FROM
  `bigquery-public-data.chicago_crime.crime`
LIMIT 5
"""

# Run the query
query_job = client.query(query)
rows = query_job.result()

# Convert to a list of dictionaries for easier display
data = []
for row in rows:
    data.append(dict(row.items()))

# Print the results
import pandas as pd
df = pd.DataFrame(data)

print(df)
```
